# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고','능남고':'상원고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] =  df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
df['결과'] = 'Fail' #결과 Column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
#총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df.loc[df['총합']>400, '결과'] = 'Pass'
df
```

## Column 삭제

```python
df.drop(columns=['총합']) #총합 Column 삭제
```

```python
df.drop(columns=['국어', '영어', '수학']) #국어, 영어, 수학 column 삭제
```

```python
df
```

## Cell 수정

```python
#4번 학생의 SW특기 데이터를 Python으로 변경
df.loc['4번', 'SW특기'] = 'Python'
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
#맨 뒤에 있는 결과 column을 앞으로 가져오고, 나머지 column 들과 합쳐서 순서 변경
df = df[[cols[-1]] + cols[0:-1]]
df
```
