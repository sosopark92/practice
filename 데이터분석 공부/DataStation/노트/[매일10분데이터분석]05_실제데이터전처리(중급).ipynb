```python
df2 = pd.DataFrame(df1.iloc[:,6:-2].drop(columns=['Unnamed: 7','판매']).set_index(['제품명']).stack()).reset_index()
df2.rename(columns= {'level_1':'일자', 0:'제고량'})
```

```python
df3 = df2.rename(columns= {'level_1':'일자', 0:'제고량'})
```

```python
df1.columns
```

```python
df1[['구분',0,'카테고리명','자재그룹','자재그룹명','제품코드','제품명','안전재고',' 분류']]
```

```python
df4 = df1[['구분',0,'카테고리명','자재그룹','자재그룹명','제품코드','제품명','안전재고',' 분류']]
```

```python
pd.merge(df4,df3,on='제품명',how='left')
```

```python
df5 = pd.merge(df4,df3,on='제품명',how='left')
```

```python
df5['공급월'] = df5['일자'].dt.month
df5['공급연도'] = df5['일자'].dt.year
df5['공급주차'] = df5['일자'].dt.week
df5['공급일자'] = df5['일자'].dt.day
```

```python
df5.head(2)
```

```python
pd.pivot_table(data=df5, index=['공급연도','공급주차'],values='제고량',aggfunc='sum')
```

```python
pd.pivot_table(data=df5, index=['공급연도','공급주차'],values='제고량',aggfunc='sum').reset_index()
```
