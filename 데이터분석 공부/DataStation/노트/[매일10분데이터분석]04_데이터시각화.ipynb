# Seaborn 

* Matplot에 기반한 Python 시각화 라이브러리 
* 통계 그래프를 그리기 위한 인터페이스 제공 
* Pandas 구조와 매우 밀접하게 연결되어 있다 (Pandas 와 매우 잘 맞는다) 

```python
import seaborn as sns 
```

```python
df1 = pd.read_csv('Data03.csv')
df1.head()
```

```python
df1.info()
```

```python
df1.describe()
```

* 범주형 

```python
sns.countplot(data=df1, x='type_of_contract2')
```

```python
sns.countplot(data=df1, x='type_of_contract2',hue='product')
```

* 연속형 

```python
sns.histplot(data=df1, x='age')
```

* 다변수 (연속/연속)

```python
sns.scatterplot(data=df1, x='age',y='Mileage')
```

```python
sns.lmplot(data=df1, x='age',y='Mileage')
```

```python
sns.jointplot(data=df1, x='age',y='Mileage')
```

```python
sns.jointplot(data=df1, x='age',y='Mileage',hue='product')
```

```python
sns.pairplot(df1[['age','Mileage','amount']])
```

```python
sns.pairplot(df1[['age','Mileage','amount','product']],hue='product')
```

* 다변수 (범주/연속)

```python
sns.barplot(data=df1, x='product',y='age')
```

```python
sns.barplot(data=df1, x='product',y='age',ci=None)
```

```python
sns.barplot(data=df1, x='product',y='amount',ci=None,estimator=sum)
```

* 다변수 (시간/연속)

```python
df1['datetime(timestmp)'] = (pd.to_datetime(df1['datetime']))
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount')
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount',ci=None)
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount',ci=None,estimator=sum)
```

# Matplot Library

* 2003년에 개발된 MATLAB과 같이 사용할 수 있는 Python 그래프 출력 라이브러리

* 디테일한 옵션을 적용할 수 있고, 다양한 형태로 그래프 출력 가능 

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

* 매직 명령어 사용 : Jupyter Notebook에서 여러 옵션을 바꿔줄때 사용

```python
#매직 명령어, Jupyter Notebook에서 여러 옵션을 바꿔줄때 사용
%matplotlib 
```

```python
# 코드 셀 바로 아래 그래프를 출력 
%matplotlib  inline
```

```python
df1 = pd.read_csv('Data04.csv')
df1.head()
```

```python
df1.info()
```

```python
df1['주문일(datetime)'] = pd.to_datetime(df1['주문일'],format='%Y%m%d')
```

```python
df2 = df1.set_index('주문일(datetime)')
df2.head()
```

```python
# 그래프 출력 
plt.plot(df2['상품구매금액'])

# 그래프를 출력 
plt.show()
```

```python
# 인터렉티브 그래프 출력 
%matplotlib tk
```

```python
plt.plot(df2['상품구매금액'])
# 그래프를 출력 
plt.show()
```

* 층별화를 통한 시각화 

```python
%matplotlib  inline
```

```python
# 그래프 출력 
plt.plot(df2['상품구매금액'],label='상품구매금액')

# 타이틀과 레이블 설정 
plt.xlabel('일자')
plt.ylabel('판매 금액')
plt.title('기간 별 상품 구매금액 추이 ')

# 그래프를 출력 
plt.show()
```

* 그래프 사이즈 조절 

```python
# 그래프 사이즈 조절
plt.figure(figsize = [10,5])

# 그래프 출력 
plt.plot(df2['상품구매금액'],label='상품구매금액')

# 타이틀과 레이블 설정 
plt.xlabel('일자')
plt.ylabel('판매 금액')
plt.title('기간 별 상품 구매금액 추이 ')

# 그래프를 출력 
plt.show()
```

* 축 설정 

```python
# 그래프 사이즈 조절
plt.figure(figsize = [10,5])

# 그래프 출력 
plt.plot(df2['상품구매금액'],label='상품구매금액')

# 축 설정
plt.xlim(['2017-01-01','2018-12-31'])
plt.ylim([0,2000000])

# 타이틀과 레이블 설정 
plt.xlabel('일자')
plt.ylabel('판매 금액')
plt.title('기간 별 상품 구매금액 추이 ')

# 그래프를 출력 
plt.show()
```
