```python
import numpy as np
import pandas as pd
```

```python
left1 = pd.DataFrame({'key':['a','b','a','a','b','c'],
                      'value':range(6)})
right1 = pd.DataFrame({'group_val':[3.5,7]}, index=['a','b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True) #merge 기본 교집합
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer') #how='outer', 합집합
```

```python
lefth = pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],
                     'key2':[2000, 2001, 2002, 2001, 2002],
                     'data':np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6,2)),
                     index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
                           [2001,2002,2000,2000,2001,2002]],
                     columns=['event1','event2'])
```

```python
lefth
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.],[3.,4.],[5.,6.]],
                    index=['a','c','e'],
                    columns=['Ohio','Nevada'])
right2 = pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[13,14]],
                     index=['b','c','d','e'],
                     columns=['Missouri','Alabama'])
```

```python
left2
```

```python
right2
```

```python
left2.join(right2,how='outer')
```

```python
left1.join(right1, on='key')
```

```python
another=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[16.,17.]],
                    index=['a','c','e','f'],
                    columns=['New York','Oregon'])
another
```

```python
s1 = pd.Series([0,1], index=['a','b'])
s2 = pd.Series([2,3,4], index=['c','d','e'])
s3 = pd.Series([5,6], index=['f','g'])
```

```python
pd.concat([s1,s2,s3]) #Series
```

```python
pd.concat([s1,s2,s3], axis=1) #DataFrame
```

```python
s4 = pd.concat([s1,s3])
```

```python
s4
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1,s4],axis=1,join='inner')
```

```python
index = ['a','c','b','e']
pd.concat([s1,s4],axis=1).reindex(index)
```

```python
result = pd.concat([s1,s2,s3], keys=['one','two','three']) #이어붙인 series 구분할 수 있도록 계층적 색인 생성
```

```python
result
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3,2), index=['a','b','c'],
                  columns=['one','two'])
df2 = pd.DataFrame(np.arange(4).reshape(2,2), index=['a', 'c'],
                  columns=['three','four'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2], axis=1, keys=['level1','level2'])
```

```python
pd.concat([df1,df2], axis=1, keys=['level1','level2'], names=['upper','lower'])
```
